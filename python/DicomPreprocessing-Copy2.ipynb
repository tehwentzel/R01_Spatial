```python
def get_finished_pids(root='../data/DICOMs/ProcessedPatients/'):
    files = glob.glob(root + 'pcloud_*.json')
    pids = []
    for file in files:
        pid = file.replace( root+'pcloud_','').replace('.json','')
        if pid.isnumeric():
            pids.append(int(pid))
        else:
            print('bad pid',pid)
    return pids

len(get_finished_pids())
```

```python
def get_all_dicom_ids(root = '../data/DICOMs/R01/'):
    files = glob.glob(root+'*/')
    ids = []
    for f  in files:
        pid = f.replace(root,'').replace('/','')
        if pid.isnumeric():
            ids.append(int(pid))
        else:
            print('bad pid',pid)
    return ids
get_all_dicom_ids()
```

```python
def load_patient_folder(root,pid,as_dict=True):
    files = glob.glob(root+str(pid) + '/**/**')
    entry = {}
    pids = set([])
    pid2 = False
    for f in files:
        p = pydicom.dcmread(f)
        pid2 = str(p.PatientName)
        pids.add(pid2)
        dtype = str(p.SOPClassUID.name)
        subentry = entry.get(dtype,[])
        subentry.append(p)
        entry[dtype] = subentry
    if len(pids) > 1:
        print('mutilple ids',pids)
    if as_dict:
        return {int(pid2): entry}
    return entry, pid2

def sample_pids(sample_size=10,skip_finished=True):
    potential_ids = get_all_dicom_ids()
    if skip_finished:
        to_skip = get_finished_pids()
        potential_ids = [pid for pid in potential_ids if pid not in to_skip]
    if sample_size is not None and sample_size < len(potential_ids):
        potential_ids = potential_ids[:sample_size]
    return potential_ids

def load_dicoms(skip_finished=True,sample_size=2):
    root = '../data/DICOMs/R01/'
    to_skip = []
    potential_ids = sample_pids(sample_size=sample_size,skip_finished=skip_finished)
    dicom_files = {}
    files = []
    for pid in potential_ids:
        pentry,new_pid = load_patient_folder(root,pid,as_dict=False)
        if str(new_pid) != str(pid):
            print('id mismatch',pid,new_pid)
        dicom_files[int(new_pid)] = pentry
    print(len(dicom_files))
    return dicom_files

dicom_files = load_dicoms(True)
dicom_files.keys()
```

```python
def get_element(ds,string,default=False):
    try:
        return ds.data_element(string).value
    except:
        return default
    
def fix_roi_name(roi):
    roi = roi.lower()
    if 'gtv' in roi:
        if 'gtvn' in roi or 'node' in roi or 'node' in roi:
            return 'gtvn'
        else:
            return 'gtv'
    if 'ptv' in roi:
        return 'ptv'
    if 'ctv' in roi:
        return 'ctv'
    return Const.organ_associations.get(roi,roi)

def read_rt_struct(rtstruct,contour_dict = None,associations=None):
    #this should read an rtstruct file, clean the names
    #returns a dict of {roi: [pointcloud,pointclouds...]}
    #multiple pointclouds if there are different contours that are name varaints of a single organ (list gtv)
    
    #pass contour dict if there are mutliple rt struct files?
    if contour_dict is None:
        contour_dict = {}
    if associations is None:
        associations=Const.organ_associations
        

    rseq_list = get_element(rtstruct,'ROIContourSequence',[])
    roi_list = get_element(rtstruct,'StructureSetROISequence',[])
    if len(rseq_list) < 1 or len(roi_list) < 1:
        return False
    assert(len(rseq_list) == len(roi_list))
    for rcseq,roi in zip(rseq_list,roi_list):
        try:
            name = roi.ROIName
            number = roi.ROINumber
            cs = rcseq.ContourSequence
            #each contourSequence is at a different z-height, so Imma just merge them
            contours = [np.array(s.ContourData).reshape(-1,3) for s in rcseq.ContourSequence if len(s.ContourData) > 0]
            contours = np.vstack(contours)
            curr_entry = contour_dict.get(name,[])
            curr_entry.append(contours)
            contour_dict[name] = curr_entry
        except Exception as e:
            print('error in read_rt_struct',e)
    return contour_dict

# rt_struct_file = dicom_files[3194171650]['RT Structure Set Storage'][0]
read_rt_struct(rt_struct_file)
```

```python
class DicomProcessor():

    
    def __init__(self,dicom_files = None, path = None, max_count = 10):
        if dicom_files is None:
            dicom_files = self.read_dicoms(path)
        self.dicoms=dicom_files    
        self.uids = list(dicom_files.keys())
        self.active_uid =  self.uids[0]
        
    def read_dicoms(self,path,max_count=None):
        files = glob.glob(path)
        dicom_files = {}
        for f in files:
            p = pydicom.dcmread(f)
            pid = str(p.PatientName)
            entry = dicom_files.get(pid,{})
            dtype = str(p.SOPClassUID.name)
            if dtype not in ['RT Dose Storage','CT Image Storage','RT Structure Set Storage']:
                continue
            subentry = entry.get(dtype,[])
            subentry.append(p)
            entry[dtype] = subentry
            dicom_files[pid] = entry
            if max_count is not None and len(dicom_files) >= max_count:
                break
        return dicom_files
    
    def get_patient(self,uid=None):
        if uid is None:
            uid = self.active_uid
        return self.dicoms.get(uid)
    
    def get_all_patients(self):
        plist = [self.dicoms.get(uid) for uid in self.uids]
        return plist
    
def get_dicom_spatial(ds):
    position = get_element(ds,'ImagePositionPatient')
    orientation = get_element(ds,'ImageOrientationPatient')
    slice_thickness = get_element(ds,'SliceThickness')
    loc = get_element(ds,'SliceLocation')
    spacing = get_element(ds,'PixelSpacing')
    data = ds.pixel_array
    entry = {
        'data': data,
        'position': position,
        'orientation': orientation,
        'location': loc,
        'spacing': spacing,
        'slice_thickness': float(slice_thickness) if slice_thickness else -1,
    }
    return entry

def get_dicom_identifiers(ds):
    #these should alway be present
    study_uid = ds.StudyInstanceUID
    series_uid = ds.SeriesInstanceUID
    class_uid = str(ds.SOPClassUID.name) #shouldn't be optional
    pid = str(ds.PatientID)
    entry = {
        'patient_id': pid,
        'study_uid': study_uid,
        'series_uid': series_uid,
        'type': class_uid,
    }
    return entry

def check_consistency(ct_images,identifiers):
    identifiers = get_dicom_identifiers(ct_images[0]) if identifiers is None else identifiers
    #just check for no
    flag = True
    for ct in ct_images:
        temp_identifiers = get_dicom_identifiers(ct)
        for k,v in temp_identifiers.items():
            if 'series' not in k and 'type' not in k and identifiers[k] != v:
                print(k,v,identifiers[k])
                flag = False
    return flag

def process_patient(pdict):
    ct_images = pdict.get('CT Image Storage')
    dose_data = pdict.get('RT Dose Storage')
    rt_structs = pdict.get('RT Structure Set Storage')
    entry = get_dicom_identifiers(ct_images[0])
    check_consistency(ct_images,entry)
    check_consistency(dose_data,entry)
    check_consistency(rt_structs,entry)
    ct_spatial = []
    for cti in ct_images:
        spatial = get_dicom_spatial(cti)
        ct_spatial.append(spatial)
        
    stacked_images = get_stacked_images(ct_spatial)

    dose_spatial = []
    for dsi in dose_data:
        spatial = get_dicom_spatial(dsi)
        dose_spatial.append(spatial)
            
    topx = lambda array: np.min([d['position'][0] for d in array])
    topy = lambda array: np.min([d['position'][1] for d in array])
    topz = lambda array: np.min([d['position'][2] for d in array])
    ct_corner = (topx(ct_spatial),topy(ct_spatial),topz(ct_spatial))
    dose_corner = (topx(dose_spatial),topy(dose_spatial),topz(dose_spatial))
    #just check for no
    ct_entry = {
        'location': ct_corner,
        'spacing': ct_spatial[0]['spacing'],
        'orientation': ct_spatial[0]['orientation'],
        'thickness': ct_spatial[0]['slice_thickness'],
        'data': stacked_images,
        'data_shape': stacked_images.shape,
    }
    
    dose_data = dose_spatial[0]['data']
    #idk if this works yet, but they mentioned summing up voxels for muliple dose files
    #i dont see that in the current data so idk
    if len(dose_spatial) > 1:
        for d in dose_spatial[1:]:
            dose_data = dose_data + d['data']
            
    dose_entry = {
        'location': dose_corner,
        'spacing': dose_spatial[0]['spacing'],
        'orientation': dose_spatial[0]['orientation'],
        'thickness': dose_spatial[0]['slice_thickness'],
        'data': dose_data,
        'data_shape': dose_data.shape
    }
    entry['ct'] = ct_entry
    entry['dose'] = dose_entry
    if np.sum(np.array(ct_entry['orientation'])-np.array(dose_entry['orientation'])) != 0:
        print('unaligned orientations',entry['patient_id'],ct_entry['orientation'],dose_entry['orientation'])
    return entry 

def get_stacked_images(ct_dicts):
    ct_dicts = sorted(ct_dicts, key = lambda x: x['location'])
    imgs = [c['data'] for c in ct_dicts]
    stack = np.stack(imgs,axis=0)
    return stack

def process_dicoms_spatial(dicom_files=None,path='../data/sample_dicoms/**/**/**'):
    dp = DicomProcessor(dicom_files=dicom_files,path=path)
    plist = [process_patient(p) for p in dp.get_all_patients()]
    pdict = {e['patient_id']: e for e in plist}
    return pdict

# pdict_spatial = process_dicoms_spatial(dicom_files=load_patient_folder('../data/DICOMs/R01/',3045556976,as_dict=True))
```

```python
def rescale_3d(image,size,categorical=None):
    if categorical is None:
        categorical = image.dtype.kind == 'i'
    image_interpolation = 'nearest' if categorical else 'linear'
    #size[::-1] is because we assume array is actually depth,x,y like in an image stack so we inverse the list
    #this sort of also assumes only z varies but whatevs this wont ever come back to haunt me
    resize = tio.Resample(size[::-1],image_interpolation = image_interpolation)
    return resize(np.expand_dims(image,0))[0]

test = (100*np.random.random((100,100,100)))
rescale_3d(test,[2,2,1]).shape
```

```python
def points_to_spatial(points, location, gridsize):
    scale = np.array(gridsize)
    corner = np.array(location)
    newpoints = points*scale + corner
    return  newpoints

def get_mask_pointclouds(pdict,scale=None):
    clouds = {}
    mask = pdict['mask']
#     image_stack = pdict['ct']['data']
    image_stack = pdict['ArrayDicom']
    roi_map = pdict['roi_mask_map']
    all_points = []
    if scale is not None:
        print('mask_pointclouds pre shapes',image_stack.shape,mask.shape)
        image_stack = rescale_3d(image_stack,scale,categorical=False)
        mask = rescale_3d(mask,scale,categorical=True)
        print('mask_pointclouds post shapes',image_stack.shape,mask.shape)
    for index, roi in roi_map.items():
        #should be x, y, z if I tranpose it (default it's slicexheightxwidth)
        contour = np.argwhere(mask.T == index)
        (cx, cy, cz) = np.where(mask.T == index)
        point_values = image_stack.T[cx,cy,cz].ravel()
        #skip bad values with < 4 points because the algorithms break
        if contour.shape[0] > 3:
            all_points.append(contour)
            entry = {'points': contour, 'values': point_values}
            clouds[roi] = entry
    
    spacing = ct['spacing']
    location = ct['location']
    thickness = ct['thickness']
    grid = np.array([spacing[0],spacing[1],thickness])*scale
    for roi, entry in clouds.items():
        coords = points_to_spatial(entry['points'],location,grid)
        entry['coordinates']= coords
        clouds[roi] = entry
    
    return clouds

def get_dose_pointcloud(pdict,threshold=100):
    dose = pdict['dose']
    data = dose['data']
    contour = np.argwhere(data.T > threshold)
    (cx,cy,cz) = np.where(data.T > threshold)
    point_values = data.T[cx,cy,cz].ravel()
    return {'points': contour, 'values': point_values}
    
def dose_pcloud(pdict,scale=None,**kwargs):
    cloud = get_dose_pointcloud(pdict,**kwargs)
    dose = pdict['dose']
    spacing = dose['spacing']
    location = dose['location']
    thickness = dose['thickness']
    grid =np.array([spacing[0],spacing[1],thickness])
    if scale is not None:
        grid = grid*scale
    coords = points_to_spatial(cloud['points'],location,grid)
    cloud['coordinates'] = coords
    return cloud

def interpolate_dose_values(roicloud,dosecloud,c=1.3):
    #calculate the dose for a pixel based on the nearest dose voxel
    #if c > 1, it takes all points within c times the distance to the closest voel
    roi_points = roicloud['coordinates']
    dose_points = dosecloud['coordinates']
    dose_values = dosecloud['values']
    interpolated_vals = np.zeros(roicloud['values'].shape)
    for i,point in enumerate(roi_points):
        pdists = cdist([point],dose_points)
        nearest = np.flatnonzero(pdists < c*pdists.min())
        dval = dose_values[nearest].mean()
        interpolated_vals[i] = dval
    return interpolated_vals

def get_cloud_worker(args):
    (roi,roicloud,dose_cloud) = args
    newcloud = {'coordinates': roicloud['coordinates']}
    dose_vals = interpolate_dose_values(roicloud,dose_cloud)
    newcloud['dose_values'] = dose_vals
    return (roi,newcloud)

def get_clouds(pdict,parallel=True,roi_scale=None,dose_scale=None):
    roi_clouds = get_mask_pointclouds(pdict,scale=roi_scale)
    dose_cloud = dose_pcloud(pdict,scale=dose_scale)
    if parallel:
        newclouds = joblib.Parallel(n_jobs=-2)(joblib.delayed(get_cloud_worker)((k,v,dose_cloud)) for k,v in roi_clouds.items())
    else:
        newclouds = [get_cloud_worker((k,v,dose_cloud)) for k,v in roi_clouds.items()]
    newclouds = {x[0]:x[1] for x in newclouds}
    return newclouds
```

```python
def pc_dist_worker(args):
    #wrapper to help parallelize pointcloud distance calclations
    if args[0] == args[1]:
        return 0
    clouds = args[2]
    return pointcloud_distance(clouds[args[0]]['coordinates'],clouds[args[1]]['coordinates'])

def roi_clouds_to_concave_hull(roi_clouds,**kwargs):
    #wrapper for taking a dict of {roi: pointcloud} and returning {roi: pointcloud_concave_hull}
    temp = {}
    for k,entry in roi_clouds.items():
        #skip if too few points 
        coords = entry['coordinates']
        values = entry['dose_values']
        if coords.shape[0] > 10:
            hull_points = concave_hull_3d(coords,return_vertices=True,**kwargs)
            temp[k] = {'coordinates': coords[hull_points], 'dose_values': values[hull_points]}
        else:
            temp[k] = entry
    return temp

def get_interorgan_distances(roi_clouds, 
                             roi_map=None, 
                             concavify=False,
                             parallel=True,
                             **kwargs):
    #calc interorgan distances on a dict of {roi: pointcloud, roi2: pointcloud2} as an array
    #also returns the order of rois in the dict [r1, r2, r3,...] =>[ [d(r1,r1),d(r1,r2),d(r1,r3)], [d(r2,r1),d(r2,r2),d(r2,r3)],...]
    if roi_map is not None:
        rois = Const.organ_list + ['gtv','gtvn','ctv','ptv']
    else:
        rois = list(roi_clouds.keys())
    valid_rois = list(roi_clouds.keys())
    rois = [r for r in rois if r in valid_rois]
    dists = {}
    #get convex hull for interorgan_distance if I remove that from the default processing
    if concavify:
        roi_clouds = roi_clouds_to_concave_hull(roi_clouds,alpha_scale=.5,**kwargs)
    for roi in ['gtv','gtvn']:
        if roi not in valid_rois:
            continue
        if parallel:
            dlist = joblib.Parallel(n_jobs=-2)(joblib.delayed(pc_dist_worker)((roi,r2,roi_clouds)) for r2 in rois)
        else:
            dlist = [pc_dist_worker((roi,r2,roi_clouds)) for r2 in rois]     
        dists[roi] = dlist
    return dists, rois

def np_converter(obj):
    #converts stuff to vanilla python  for json since it gives an error with np.int64 and arrays
    if isinstance(obj, np.integer):
        return int(obj)
    elif isinstance(obj, np.float):
        return round(float(obj),3)
    elif isinstance(obj, np.ndarray):
        return obj.tolist()
    elif isinstance(obj, np.bool_):
        return bool(obj)
    elif isinstance(obj, datetime.datetime) or isinstance(obj, datetime.time):
        return obj.__str__()
    print('np_converter cant encode obj of type', obj,type(obj))
    return obj

def pickle_individual_patient(pdict,folder=None):
    folder = '../data/DICOMs/ProcessedPatients/' if folder is None else folder
    fname = folder + 'pcloud_' + str(pdict['patient_id'])
    to_keep = ['patient_id','study_uid','series_uid','type','roi_mask_map','Volumes',
               'distances','spacing',
               'point_clouds']
    entry = {k:pdict.get(k) for k in to_keep}
    with open(fname+'.json','w') as f:
        simplejson.dump(entry,f,default=np_converter)
#     with open(fname+'.p','wb') as f:
#         pickle.dump(entry,f)
    images = {'ct': pdict['ct']['data'], 'dose': pdict['dose']['data']}
    fname2 = folder + 'images_' + str(pdict['patient_id'])
    with open(fname2+'.json','w') as f:
        simplejson.dump(images,f,default=np_converter)
#     with open(fname2+'.p','wb') as f:
#         pickle.dump(images,f)
    return True

process_patient(load_patient_folder('../data/DICOMs/R01/',3045556976,as_dict=False)[0])

def merge_dicom_dicts(pids=None,
                      pdict_spatial=None,
                      dicom_root='../data/DICOMs/R01/',
                      folder=None,
                      max_patients=50,
                      skip_finished=True,
                      scale_xy=2,
                      scale_z=1.25,
                      dose_scale=None,
                     ):
    new_entries = []
    to_copy = [
        'mask','roi_mask_map','Volumes',
        'Description','SeriesInstanceUID','StudyInstanceUID'
    ]
    if pids is None:
        if pdict_spatial is not None:
            pids = list(pdict_spatial.keys())
        else:
            pids = sample_pids(sample_size=max_patients,skip_finished=skip_finished)
    roi_scale=np.array([scale_xy,scale_xy,scale_z])
    for pid in pids:
        if pdict_spatial is None or pid not in pdict_spatial:
            temp = load_patient_folder(dicom_root,pid,as_dict=False)[0]
            if len(temp) < 1:
                print('bad id',pid)
                continue
            entry = process_patient(temp)
        else:
            entry = pdict_spatial.get(pid)
            
        if len(entry) < 1 or entry is None:
            print('bad id',pid)
            
        
        spacing = np.array(list(entry['ct']['spacing']) + [ entry['ct']['thickness']])
        print('spacing',spacing)
        entry['spacing'] = spacing
        entry['spacing_adjusted'] = spacing * roi_scale
        dose_cloud = dose_pcloud(entry,scale=dose_scale)
        
        dr = dicom_reader_from_ids([pid],root=dicom_root)
        point_clouds = {}
        for i,roilist in enumerate([Const.organ_list[:],['gtv','gtvn'],['ctv','ptv']]):
            plist, _ = get_all_patients(dr,rois= roilist)
            pentry = plist[0]
            if pentry['mask'].shape[0] != entry['ct']['data'].shape[0]:
                print('mismatched shape for pid',pid,'|',pentry['mask'].shape[0],entry['ct']['data'].shape[0])
            if i == 0:
                
            pentry['spacing'] = entry['spacing']
            tempclouds = get_clouts(pentry,dose_cloud, roi_scale=roi_scale)
            for k,v in tempclouds:
                point_clouds[k] = v
        entry = {k:v for k,v in entry.items()}
        for key in to_copy:
            entry[key] = pentry[key]
        
        
#         print(entry.keys(),entry['ct'].keys(),entry['dose'].keys())
        
#         pclouds = get_clouds(entry,roi_scale=roi_scale)
        
        entry['point_clouds'] = pclouds
        print(pid,'pointclouds done')
        distances = get_interorgan_distances(pclouds,parallel=True)
        print(pid,'distaces donse',distances)
        entry['distances'] = distances
        pickle_individual_patient(entry)
        new_entries.append(pid)
        print(pid,'done')
        print('__________')
    return new_entries

pdata = merge_dicom_dicts(max_patients=2)
pdata
```
